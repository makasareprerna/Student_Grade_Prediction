```python
# ML models for comparison purposes
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor  
from sklearn import neighbors

# Train Test split data
from sklearn.model_selection import train_test_split


# Metric to be used
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score


# Distributions
import scipy
```

```python
# Train and test data splitting into (80% and 20%)
x_train, x_test, y_train, y_test = train_test_split(math_final, math_final['G3'], test_size = 0.20, random_state=42)
```

```python
x_train.head()
```

```python
# MAE - Mean Absolute Error & RMSE - Root Mean Square Error prediction of models
```

```python
# Evaluate several ml models by training on training set and testing on testing set
def evaluate(x_train, x_test, y_train, y_test):
    # Names of models
    model_name_list = ['Linear Regression', 'ElasticNet Regression',
                      'Random Forest', 'Extra Trees', 'SVM',
                       'Gradient Boosted', 'Ridge','Lasso','Decision Tree','Knn']
    x_train = x_train.drop('G3', axis='columns')
    x_test = x_test.drop('G3', axis='columns')
    
    
    
    # Instantiate the models
    model1 = LinearRegression()
    model2 = ElasticNet(alpha=1.0, l1_ratio=0.5)
    model3 = RandomForestRegressor(n_estimators=100)
    model4 = ExtraTreesRegressor(n_estimators=100)
    model5 = SVR(kernel='rbf', degree=3, C=1.0, gamma='auto')
    model6 = GradientBoostingRegressor(n_estimators=50)
    model7 = Ridge(alpha=1.0,normalize=True)
    model8 = Lasso(alpha=1.0,normalize=True)
    model9 = DecisionTreeRegressor(random_state = 0)
    model10 = neighbors.KNeighborsRegressor(10,weights='uniform')

    
    # Dataframe for result
    result = pd.DataFrame(columns=['MAE', 'RMSE','Accuracy'], index = model_name_list)
    
    # Train and predict with each model
    for i, model in enumerate([model1, model2, model3, model4, model5, model6, model7, model8, model9 , model10]):
        model.fit(x_train, y_train)
        prediction = model.predict(x_test)
        
        # Metrics
        mae = np.mean(abs(prediction - y_test))
        rmse = np.sqrt(np.mean((prediction - y_test) ** 2))
        r2=r2_score(y_test, prediction)
        
        # Insert results into the dataframe
        model_name = model_name_list[i]
        result.loc[model_name, :] = [mae, rmse, ((abs(r2))*100)]
    
    
    return result
```

```python
result = evaluate(x_train, x_test, y_train, y_test)
result
```

```python
plt.figure(figsize=(12, 7))

# Root mean squared error
ax =  plt.subplot(1, 2, 1)
result.sort_values('MAE', ascending = True).plot.bar(y = 'MAE', color = 'blue', ax = ax)
plt.title('Model Mean Absolute Error') 
plt.ylabel('MAE')

# Median absolute percentage error
ax = plt.subplot(1, 2, 2)
result.sort_values('RMSE', ascending = True).plot.bar(y = 'RMSE', color = 'blue', ax = ax)
plt.title('Model Root Mean Squared Error') 
plt.ylabel('RMSE')

plt.show()
```
